```python
# Ваш код здесь
def winers(scores: dict):
    scoreslist = sorted(scores.values(), reverse=True)
    final = 1
    for i in scoreslist:
        for k, v in scores.items():
            if v == i:
                print(f'{final} место - {k}: {i}')
        final += 1
        if final > 3:
            break
```

Дан список кортежей `constants`. В каждом кортеже на первом месте лежит обозначение математической или физической константы, а на втором месте — её значение, например: ('g', '9.823773828293883939291').

Определите функцию, которая из списка кортежей преобразует данные в словарь, в котором обозначение константы — это ключ, а соответствующее число — значение.

Для выполнения этого задания воспользуйтесь методом `get` для словарей.


Пример: 

Output: 

`Введите имя константы:`

`pi`

`pi = 3.14159265358979323846264338327950288`

Подсказка

Метод `get` для `dict` ищет значение по указанному ключу в исходном словаре. Если значение не найдено, то возвращает сообщение, указанное вторым аргументом.

Пример:

```python
d = {1: '1', 2: '2'}
print(d.get(1, 'не найдено'))
print(d.get(3, 'не найдено'))
```

```python
constants = [('g', '9.823773828293883939291'), ('pi', '3.14159265358979323846264338327950288'), \
             ('L', '0.5'), ('e', '2.71828182845904523536028747135266249'), \
             ('B2', '1.902160583104'), ('C2', '0.64341054629')]
```

```python
# Ваш код здесь
def list_2_dict(list1):
    constants_dict = {}
    for i in list1:
        constants_dict[i[0]] = i[1]
    return constants_dict
```

```python
constants_dict = list_2_dict(constants)
```

```python
key = input('Введите имя константы: ')
print(f'{key} = {constants_dict.get(key, "не найдено")}')
```

*Задача 6. Закупки*

Дан список `orders`, который состоит из кортежей. Каждый кортеж — это закупка определённого покупателя. В нём хранится следующая информация:

(id покупателя, количество товаров, сумма покупки)

Определите размер скидки для каждого покупателя по истории покупок из списка `orders`. Рассчитать скидку можно по следующему алгоритму:
- если сумма покупок от 10 тыс руб — скидка 5%;
- если сумма покупок от 15 тыс руб — скидка 10%;
- если сумма покупок от 20 тыс руб — скидка 15%;
- если сумма покупок менее 10 тыс руб — скидка 0%.

Результаты расчёта запишите в словарь `dicsount_system` в следующем виде:

- ключ — id покупателя;
- значение — словарь вида: {'discount': 5, 'total_sum': 12000}, где `discount` — размер скидки, `total_sum` — общая сумма покупок.

Используйте функции и декомпозицию.


```python
# Пример результата
```

```python
orders = [
          (111, 3, 1200), (118, 9, 8000),
          (111, 10, 1000000), (119, 9, 13300),
          (156, 10, 11000), (123, 9, 11200),
          (178, 20, 3200), (178, 5, 8900),
          (156, 9, 8000), (123, 4, 3500),
          (101, 9, 8000), (101, 8, 1000),
]
```

```python
# Ваш код здесь
def orders_2_dict(orders):
    orders_dict = {}
    for order in orders:
        orders_dict[order[0]] = {}
        orders_dict[order[0]]['discount'] = 0
        orders_dict[order[0]]['total_sum'] = 0
    return orders_dict

def insert_sum(orders, orders_dict):
    for order in orders:
        orders_dict[order[0]]['total_sum'] += order[2]
    return orders_dict

def calc_discount(orders_dict):
    for order in orders_dict.values():
        if order['total_sum'] > 20000:
            order['discount'] = 15
        elif order['total_sum'] > 15000:
            order['discount'] = 10
        elif order['total_sum'] > 10000:
            order['discount'] = 5
    return orders_dict

orders_dict = orders_2_dict(orders)
orders_dict = insert_sum(orders, orders_dict)
discount_system = calc_discount(orders_dict)
for k, v in discount_system.items():
    print(k, v)
```

Далее напишите программу, которая будет запрашивать id пользователя и сумму его покупки. По данным скидок из словаря `discount_system` рассчитайте сумму покупки со скидкой и выведите на экран. После обновите значения `discount` и `total_sum` по данному пользователю. 

Если скидка изменилась, выведите сообщение о его новой скидке на последующие покупки.

Если пришёл новый пользователь, просто добавьте его данные в словарь `discount_system`.


```python
def add_id(user_id, summ):
    global discount_system
    discount_system[user_id] = {'discount': 0, 'total_sum': summ}
    
def add_summ(user_id, summ):
    global discount_system
    discount_system[user_id]['total_sum'] += summ

def recalc_discount(user_id):
    global discount_system
    old_discount = discount_system[user_id]['discount']
    if discount_system[user_id]['total_sum'] > 20000:
        discount_system[user_id]['discount'] = 15
    elif discount_system[user_id]['total_sum'] > 15000:
        discount_system[user_id]['discount'] = 10
    elif discount_system[user_id]['total_sum'] > 10000:
        discount_system[user_id]['discount'] = 5
    return old_discount != discount_system[user_id]['discount']

orders_dict = orders_2_dict(orders)
orders_dict = insert_sum(orders, orders_dict)
discount_system = calc_discount(orders_dict)
for k, v in discount_system.items():
    print(k, v)

while True:
    user_id = int(input('ID?: '))
    if not user_id:
        break
    summ = int(input('Стоимость покупки?: '))
    if user_id in discount_system:
        print(f"стоимость покупки с {discount_system[user_id]['discount']} = {summ * (100-discount_system[user_id]['discount'])/100}")
        add_summ(user_id, summ)
        if recalc_discount(user_id):
            print(f"Ваша программа лояльности обновилась, новая скидка: {discount_system[user_id]['discount']}")

    else:
        add_id(user_id, summ)
```
