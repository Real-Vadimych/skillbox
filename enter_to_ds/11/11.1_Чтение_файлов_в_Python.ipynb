В качестве практического занятия предлагаем вам усложнить задачу и подсчитать число уникальных слов в рассказе Чехова. Для этого вам нужно читать файл слово за словом, как мы делали в уроке, но добавлять слова не в список, а в множество (set). Множество само удалит дубликаты, и вы получите список уникальных слов. Вам останется только вывести размер множества на экран.

Проверьте себя! Правильный ответ: 1198.

```python
words = set()
with open('anna_words.txt') as file:
    for word in file:
        words.add(word)
        
print(len(words))
```
